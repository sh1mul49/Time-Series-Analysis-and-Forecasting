## Data loading

Load the train and test data and show their dimension and top 5 rows .


```python
import pandas as pd
train_df = pd.read_excel('DataTrain.xlsx')
test_df = pd.read_excel('DataTest.xlsx')
print(train_df.shape, test_df.shape)
display(train_df, test_df)
```

Understand the range of train and test data

```python
train_df['Date Time'].min(), train_df['Date Time'].max()
```

```python
test_df['Date Time'].min(), test_df['Date Time'].max()
```

Combine train and test data for basic data preprocessing

```python
df = pd.concat([train_df, test_df], ignore_index=True)
print(df.shape)
df
```

# Dealing with missing values with Interpolation

Interpolation is a method used to estimate missing values in data based on the known surrounding values. If a value is missing between two known values, estimate it by looking at the values before and after.

```python
df.isnull().sum()
```

```python
sensor_cols = [col for col in df.columns if 'Avg_' in col]
df[sensor_cols] = df[sensor_cols].interpolate(method='linear', limit_direction='both')
```

```python
df.isnull().sum()
```

Identify columns that have whitespace or special characters and remove them

```python
print([col for col in df.columns if not col.isidentifier()])
```

```python
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
#Sort by Data Time for the purpose of compute rolling average by considering near DateTime response
df = df.sort_values(by=['Date Time'])
```

```python
# compute rolling inj_diff for filling the missing value in the df which would help for rolling featues creation later for test data
rolling_avg = df['inj_diff'].rolling(window=24, min_periods=1).mean()
df['inj_diff'] = df['inj_diff'].fillna(rolling_avg)
df
```

Date Related feature creation from Date Time variable

```python

df['hour'] = df['Date Time'].dt.hour
df['day_of_week'] = df['Date Time'].dt.dayofweek
df['month'] = df['Date Time'].dt.month
df['day_of_year'] = df['Date Time'].dt.dayofyear
df['week_of_year'] = df['Date Time'].dt.isocalendar().week
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
```

Split the combined dataframe back into train and test sets based on the 'Rand' column and perform further preprocessing seperatly to get rid of data lekage

```python

train_data = df[df['Rand'].isin(train_df['Rand'])]
test_data = df[df['Rand'].isin(test_df['Rand'])]
test_data = test_data.rename(columns={'inj_diff': 'Computed_inj_diff_RollingMethod'})
# Verify the shapes of the seperated dataframes with the loaded train_df and test_df
print(train_df.shape, test_df.shape)
print(train_data.shape, test_data.shape)

display(train_data, test_data)
```

Featue Creation through Lag and Rolling average of previous CO2 injection rate for the train data

```python
# The value 1 refers to 1-hour lag
train_data['lag_1'] = train_data['inj_diff'].shift(1)
train_data['lag_2'] = train_data['inj_diff'].shift(2)
train_data['lag_3'] = train_data['inj_diff'].shift(3)
train_data['lag_1D'] = train_data['inj_diff'].shift(24)
train_data['lag_2D'] = train_data['inj_diff'].shift(48)
train_data['lag_3D'] = train_data['inj_diff'].shift(72)

# Rolling mean features. window represnts the number of previus value will be considerd and min_periods represnt min number of value require to create the featues
train_data['rolling_mean_2'] = train_data['inj_diff'].rolling(window=2,min_periods=1).mean()
train_data['rolling_mean_3'] = train_data['inj_diff'].rolling(window=3,min_periods=1).mean()
train_data['rolling_mean_4'] = train_data['inj_diff'].rolling(window=4,min_periods=1).mean()
train_data['rolling_mean_5'] = train_data['inj_diff'].rolling(window=5,min_periods=1).mean()
train_data['rolling_mean_6'] = train_data['inj_diff'].rolling(window=6,min_periods=1).mean()
train_data['rolling_mean_7'] = train_data['inj_diff'].rolling(window=7,min_periods=1).mean()
train_data['rolling_mean_8'] = train_data['inj_diff'].rolling(window=8,min_periods=1).mean()
train_data['rolling_mean_9'] = train_data['inj_diff'].rolling(window=9,min_periods=1).mean()
train_data['rolling_mean_10'] = train_data['inj_diff'].rolling(window=10,min_periods=1).mean()

# Rolling mean over the past 3 days (72 hours)
train_data['rolling_mean_3_days'] = train_data['inj_diff'].rolling(window=72,min_periods=1).mean()

# Rolling max/min over the past 7 days (168 hours)
train_data['rolling_max_7'] = train_data['inj_diff'].rolling(window=168,min_periods=1).max()
train_data['rolling_min_7'] = train_data['inj_diff'].rolling(window=168,min_periods=1).min()
```

Featue Creation through Lag and Rolling average of previous CO2 injection rate for the Test Data

```python
#The value 1 refers to 1-hour lag
test_data['lag_1'] = test_data['Computed_inj_diff_RollingMethod'].shift(1)
test_data['lag_2'] = test_data['Computed_inj_diff_RollingMethod'].shift(2)
test_data['lag_3'] = test_data['Computed_inj_diff_RollingMethod'].shift(3)
test_data['lag_1D'] = test_data['Computed_inj_diff_RollingMethod'].shift(24)
test_data['lag_2D'] = test_data['Computed_inj_diff_RollingMethod'].shift(48)
test_data['lag_3D'] = test_data['Computed_inj_diff_RollingMethod'].shift(72)

# Rolling mean features. window represnts the number of previus value will be considerd and min_periods represnt min number of value require to create the featues
test_data['rolling_mean_2'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=2,min_periods=1).mean()
test_data['rolling_mean_3'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=3,min_periods=1).mean()
test_data['rolling_mean_4'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=4,min_periods=1).mean()
test_data['rolling_mean_5'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=5,min_periods=1).mean()
test_data['rolling_mean_6'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=6,min_periods=1).mean()
test_data['rolling_mean_7'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=7,min_periods=1).mean()
test_data['rolling_mean_8'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=8,min_periods=1).mean()
test_data['rolling_mean_9'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=9,min_periods=1).mean()
test_data['rolling_mean_10'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=10,min_periods=1).mean()

# Rolling mean over the past 3 days (72 hours)
test_data['rolling_mean_3_days'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=72,min_periods=1).mean()

# Rolling max/min over the past 7 days (168 hours)
test_data['rolling_max_7'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=168,min_periods=1).max()
test_data['rolling_min_7'] = test_data['Computed_inj_diff_RollingMethod'].rolling(window=168,min_periods=1).min()
```

Split train data into target(y) and features(X)

```python
X=train_final_featues.drop(columns=['inj_diff'])
y=train_final_featues['inj_diff']
```

Split the train data into train and validation using time series split since order is important for time series data

```python
from sklearn.model_selection import TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)

for train_index, val_index in tscv.split(X):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

Radom Forest Regressor Model Training and Validation  

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state = 42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
val_predictions=model.predict(X_val)
mae=mean_absolute_error(val_predictions, y_val)
rsquare_value=r2_score(val_predictions, y_val)
rmse = mean_squared_error(y_val, val_predictions) ** 0.5
print(rmse, rsquare_value,mae)
```

Extreme Gradient Boosting Model Training and Validation

```python
from xgboost import XGBRegressor
modelXGB = XGBRegressor(random_state=42)
modelXGB.fit(X_train, y_train)

xgb_val_predictions = modelXGB.predict(X_val)

maeX = mean_absolute_error(xgb_val_predictions, y_val)
rsquare_valueX = r2_score(xgb_val_predictions, y_val)
rmseX = mean_squared_error(y_val, xgb_val_predictions) ** 0.5

print(rmseX, rsquare_valueX, maeX)
```
